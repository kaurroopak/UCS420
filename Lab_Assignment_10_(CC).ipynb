Q1. Write a unique paragraph (5-6 sentences) about your favorite topic (e.g., sports,
technology, food, books, etc.).
1. Convert text to lowercase and remove punctuation using re.
2. Tokenize the text into words and sentences.
3. Split using split() and word_tokenize() and compare how Python split and NLTKâ€™s
word_tokenize() differ.
4. Remove stopwords (using NLTK's stopwords list).
5. Display word frequency distribution (excluding stopwords).

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
import matplotlib.pyplot as plt
from collections import Counter

nltk.download('punkt_tab')
nltk.download('punkt')
nltk.download('stopwords')
```

```python
text = "Artificial Intelligence (AI) is transforming many industries. It encompasses various technologies, such as machine learning, natural language processing, and robotics, which enable machines to perform tasks that typically require human intelligence. AI can be used to solve complex problems, automate tasks, and improve decision-making processes. In the future, AI may help tackle global challenges like climate change, healthcare, and education."

# Convert text to lowercase and remove punctuation
text_lower = text.lower()
text_clean = re.sub(r'[^\w\s]', '', text_lower)

# 2. Tokenize the text into words and sentences
sentences = sent_tokenize(text_clean)
words = word_tokenize(text_clean)

# 3. Compare split() and word_tokenize()
split_words = text_clean.split()
nltk_words = word_tokenize(text_clean)
print(f"Words using split(): {split_words}")
print(f"Words using word_tokenize(): {nltk_words}")

# 4. Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in nltk_words if word not in stop_words]

# 5. Display word frequency distribution
word_freq = Counter(filtered_words)
print(f"Word Frequency Distribution: {word_freq}")

# Plot the word frequency
word_freq_plot = word_freq.most_common(10)
words, freq = zip(*word_freq_plot)
plt.bar(words, freq)
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 10 Word Frequencies (Excluding Stopwords)')
plt.xticks(rotation=45)
plt.show()
```

Q3. Choose 3 short texts of your own (e.g., different news headlines, product reviews).
1. Use CountVectorizer to generate the Bag of Words representation.
2. Use TfidfVectorizer to compute TF-IDF scores.
3. Print and interpret the top 3 keywords from each text using TF-IDF.

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Short texts
texts = [
    "The AI revolution is changing the way we interact with technology. Machine learning plays a crucial role in this transformation.",
    "Blockchain technology is revolutionizing industries by providing secure and transparent transactions. It has applications in finance, healthcare, and more.",
    "Cloud computing is the future of data storage and processing. It enables scalable, flexible, and cost-effective solutions for businesses."
]

# 1. CountVectorizer (Bag of Words)
count_vectorizer = CountVectorizer()
count_vector = count_vectorizer.fit_transform(texts)
count_words = count_vectorizer.get_feature_names_out()
count_words_freq = count_vector.toarray()

# 2. TfidfVectorizer (TF-IDF)
tfidf_vectorizer = TfidfVectorizer()
tfidf_vector = tfidf_vectorizer.fit_transform(texts)
tfidf_words = tfidf_vectorizer.get_feature_names_out()
tfidf_freq = tfidf_vector.toarray()

# 3. Print top 3 keywords from each text using TF-IDF
top_n = 3
for i, text in enumerate(texts):
    text_tfidf = [(word, tfidf_freq[i][j]) for j, word in enumerate(tfidf_words)]
    sorted_tfidf = sorted(text_tfidf, key=lambda x: x[1], reverse=True)[:top_n]
    print(f"Top 3 TF-IDF keywords for Text {i+1}: {sorted_tfidf}")
```

Q4. Write 2 short texts (4â€“6 lines each) describing two different technologies (e.g., AI vs
Blockchain).
1. Preprocess and tokenize both texts.
2. Calculate:

a. Jaccard Similarity using sets

b. Cosine Similarity using TfidfVectorizer + cosine_similarity()

c. Analyze which similarity metric gives better insights in your case.

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Short texts describing different technologies
tech_texts = [
    "Artificial Intelligence (AI) uses algorithms and data to simulate human intelligence. It is widely used in applications such as natural language processing, robotics, and computer vision.",
    "Blockchain is a distributed database technology that enables secure and transparent transactions. It works by maintaining a chain of blocks, each containing transaction data."
]

# Preprocessing and Tokenization
tech_texts_clean = [re.sub(r'[^\w\s]', '', text.lower()) for text in tech_texts]

# Jaccard Similarity
set1 = set(tech_texts_clean[0].split())
set2 = set(tech_texts_clean[1].split())
jaccard_similarity = len(set1.intersection(set2)) / len(set1.union(set2))
print(f"Jaccard Similarity: {jaccard_similarity}")

# Cosine Similarity using TF-IDF
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(tech_texts_clean)
cosine_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
print(f"Cosine Similarity: {cosine_sim[0][0]}")

# Analyze the better similarity metric
print("Jaccard similarity measures the overlap of terms and is easier to interpret when comparing shorter texts. Cosine similarity is better for comparing longer and more complex texts as it takes into account the frequency of terms.")
```

Q5. Write a short review for a product or service.
1. Use TextBlob or VADER to find polarity & subjecÆŸvity for each review.
2. Classify reviews into Positive / Negative / Neutral.
3. Create a word cloud using the wordcloud library for all positive reviews.

```python
from textblob import TextBlob
from wordcloud import WordCloud

# Short product review text
reviews = [
    "This product is amazing! I love the quality and the customer service is fantastic.",
    "The product was okay, but it could be improved. Shipping was delayed, which was frustrating.",
    "I didn't like this product. It broke within a few days of use. Poor quality."
]

# Polarity and Subjectivity
for review in reviews:
    blob = TextBlob(review)
    print(f"Review: {review}")
    print(f"Polarity: {blob.sentiment.polarity}, Subjectivity: {blob.sentiment.subjectivity}")

# Classify as Positive, Negative, or Neutral
for review in reviews:
    blob = TextBlob(review)
    if blob.sentiment.polarity > 0:
        print("Positive Review")
    elif blob.sentiment.polarity < 0:
        print("Negative Review")
    else:
        print("Neutral Review")

# Word Cloud for Positive Reviews
positive_reviews = [review for review in reviews if TextBlob(review).sentiment.polarity > 0]
positive_text = ' '.join(positive_reviews)
wordcloud = WordCloud().generate(positive_text)

# Display Word Cloud
plt.figure(figsize=(8, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
!pip install tensorflow
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer # Import from tensorflow.keras.preprocessing
from tensorflow.keras.preprocessing.sequence import pad_sequences # Import from tensorflow.keras.preprocessing
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding
from tensorflow.keras.utils import to_categorical # Import from tensorflow.keras.utils
import tensorflow as tf
import random

# 1. Choose your paragraph (~100 words)
training_text = """
Artificial intelligence is transforming the world. It powers recommendation systems, chatbots, autonomous vehicles, and more.
Deep learning, a subset of AI, allows machines to recognize patterns and improve over time.
Natural language processing enables machines to understand and respond to human language.
As technology evolves, AI will become increasingly integrated into daily life, revolutionizing healthcare, education, finance, and other industries.
"""

# 2. Tokenize text using Keras Tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts([training_text])
total_words = len(tokenizer.word_index) + 1

# Create input sequences using every possible n-gram
input_sequences = []
for line in training_text.split('.'):
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i+1]
        input_sequences.append(n_gram_sequence)

# Pad sequences
max_sequence_len = max([len(x) for x in input_sequences])
input_sequences = pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre')

# Split data into input (X) and output (y)
X = input_sequences[:, :-1]
labels = input_sequences[:, -1]
y = to_categorical(labels, num_classes=total_words)

# 3. Build LSTM model
model = Sequential()
model.add(Embedding(total_words, 64, input_length=max_sequence_len - 1))
model.add(LSTM(100))
model.add(Dense(total_words, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train model
model.fit(X, y, epochs=200, verbose=0)  # You can increase epochs if needed

# 4. Generate 2â€“3 new lines of text from a seed word
seed_text = "Artificial"
next_words = 30

for _ in range(next_words):
    token_list = tokenizer.texts_to_sequences([seed_text])[0]
    token_list = pad_sequences([token_list], maxlen=max_sequence_len - 1, padding='pre')
    predicted = model.predict(token_list, verbose=0)
    predicted_word_index = np.argmax(predicted)

    for word, index in tokenizer.word_index.items():
        if index == predicted_word_index:
            seed_text += " " + word
            break

print("Generated Text:\n", seed_text)
```
